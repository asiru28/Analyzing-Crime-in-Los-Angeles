| Column     | Description              |

| `'AREA NAME'` | The 21 Geographic Areas or Patrol Divisions are also given a name designation that references a landmark or the surrounding community that it is responsible for. | `'Vict Descent'` | Victim's descent:<ul><li>`A` - Other Asian</li><li>`B` - Black</li><li>`C` - Chinese</li><li>`D` - Cambodian</li><li>`F` - Filipino</li><li>`G` - Guamanian</li><li>`H` - Hispanic/Latin/Mexican</li><li>`I` - American Indian/Alaskan Native</li><li>`J` - Japanese</li><li>`K` - Korean</li><li>`L` - Laotian</li><li>`O` - Other</li><li>`P` - Pacific Islander</li><li>`S` - Samoan</li><li>`U` - Hawaiian</li><li>`V` - Vietnamese</li><li>`W` - White</li><li>`X` - Unknown</li><li>`Z` - Asian Indian</li> |

| `'LOCATION'` | Street address of the crime. |

```python
# Re-run this cell
#Â Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
crimes = pd.read_csv("crimes.csv", parse_dates=["Date Rptd", "DATE OCC"], dtype={"TIME OCC": str})
crimes.head()
```

### Extract the first two digits from "TIME OCC", representing the hour, and convert to integer data type

```python
crimes["HOUR OCC"] = crimes["TIME OCC"].str[:2].astype(int)
```

```python
crimes.head()
```

```python
sns.countplot(data=crimes, x="HOUR OCC")
plt.show()
```

```python
# Midday has the largest volume of crime
peak_crime_hour = 12
```

```python
crimes["night_crime"] = (crimes["TIME OCC"]>="2200") | (crimes["TIME OCC"]<="0359")
```

```python
peak_night_crime_location = crimes.groupby("AREA NAME", as_index=True)["night_crime"].count().sort_values(ascending=False)
```

### The number of crimes committed against victims of different age groups.

```python
age_labels = ["0-17", "18-25", "26-34", "35-44", "45-54", "55-64", "65+"]
age_bins = [0, 17, 25, 34, 44, 54, 64, np.inf]
```

```python
crimes["Age Bracket"] = pd.cut(crimes["Vict Age"], labels=age_labels, bins=age_bins)
```

```python
sns.histplot(data=crimes, x="Vict Age", hue="Age Bracket")
plt.show()
```

```python
victim_ages = crimes["Age Bracket"].value_counts()
```
